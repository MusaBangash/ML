```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv("homeprices1.csv")
```

```python
df.head()
```

```python
%matplotlib inline
plt.scatter(df.age,df.price,color='red',marker='+')
```

```python
plt.plot(df.age,df.price,color='red')
```

```python
plt.bar(df.age,df.price)
```

```python
#There is some null value in bedroom so we can fill with different value like in this case median 

import math
median_bedrooms=math.floor(df.bedrooms.median())
```

```python
median_bedrooms
```

```python
df.bedrooms.fillna(median_bedrooms,inplace=True)
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(df[['area','bedrooms','age']],df.price,inplace=True)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[3000,4,15]])
```
